```python
import numpy as np
import pandas as pd
import re
import time
from datasketch import MinHash, MinHashLSHForest
```

```python
#Number of Permutations
permutations = 128

```

```python
minhash = []
for i in range(len(train_y)):
  m = MinHash(num_perm=permutations)
  for j in range(len(train_X.columns)):
    m.update(str(train_X.iloc[i,j]).encode('utf8'))
  minhash.append(m)

  
```

```python
forest = MinHashLSHForest(num_perm=permutations)
for i,m in enumerate(minhash):
        forest.add(i,m)
```

```python
forest.index()
result=[]
```

```python
for i in range(len(test_X)):
  m = MinHash(num_perm=permutations)
  for j in range(len(test_X.columns)):
    m.update(str(test_X.iloc[i,j]).encode('utf8'))
  id=forest.query(m,1)
  result.append(train_y[id].iloc[0])
```
